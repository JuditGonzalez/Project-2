```python
import pandas as pd
url= "https://raw.githubusercontent.com/JuditGonzalez/Project-2/main/data%20(1).csv"
df= pd.read_csv(url)
df.head()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
sns.regplot(x="danceability", y="liveness", data=df)
plt.ylim(0,)
```
